```python
X_test=X_test.str.lower()
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')  # download stopwords list
stop_words = set(stopwords.words('english'))
```

```python
type(X_train)
```

```python
X_train_clean = X_train.apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
X_test_clean = X_test.apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')  # download dictionary for lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
X_train_clean = X_train_clean.apply(lambda x: ' '.join([lemmatizer.lemmatize(word, pos='v') for word in x.split()]))
```

```python
X_test_clean = X_test_clean.apply(lambda x: ' '.join([lemmatizer.lemmatize(word, pos='v') for word in x.split()]))
```

```python
# now changeing into numerical val bying tfidf
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=3000)  # limit to top 3000 features
X_train_clean = tfidf.fit_transform(X_train_clean)  # fit and transform training data
X_test_clean = tfidf.transform(X_test_clean)  # fit and transform training data
```

```python
X_train_clean
```

```python
# dekhna h kaisa dikh rha
```

```python
X_train_dense = X_train_clean.toarray()
print(X_train_dense.shape)  # e.g., (num_samples, 3000)
print(X_train_dense[0])  
```

```python
# Create Logistic Regression model
lr_model = LogisticRegression(max_iter=1000) 

# Train the model on TF-IDF features
lr_model.fit(X_train_clean, Y_train)
```

```python
y_pred = lr_model.predict(X_test_clean)
```

```python
# Accuracy
print("Accuracy:", accuracy_score(Y_test, y_pred))

# Confusion Matrix
# print("Confusion Matrix:\n", confusion_matrix(Y_test, y_pred))

# # Detailed classification report
# print("Classification Report:\n", classification_report(Y_test, y_pred))
```

```python
# as accurancy is low as dataset is imbalance so we do ensemble learning
```

```python
# oversampling kr rhe
```

```python

oversample = RandomOverSampler()

X_train_balanced, y_train_balanced = oversample.fit_resample(X_train_clean, Y_train)
# It learns how to oversample/minority class (e.g., using SMOTE it learns the structure of the minority class)
```

```python
rf = RandomForestClassifier(n_estimators=200)
nb = MultinomialNB()
ensemble = VotingClassifier(
    estimators=[
        ('lr', lr_model),
        ('rf', rf),
        ('nb', nb)
    ],
    voting='soft'
)
```

```python
ensemble.fit(X_train_balanced, y_train_balanced)
```

```python
y_pred = ensemble.predict(X_test_clean)
```

```python

print("Accuracy:", accuracy_score(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```
